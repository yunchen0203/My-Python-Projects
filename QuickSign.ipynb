```python
%pip install pandas
import pandas as pd
from datetime import datetime
#儲存格式{學號：班級作號，姓名}
if 'member_db = {}' not in locals():
    member_db={}
else:
  print(f"繼續使用現有資料庫")
```

```python
# 準備一個 List 來存每一天的簽到紀錄
attendance_log = {}
# 要求的「規格限制」
ID_LENGTH=8
SEAT_LENGTH=5
```

```python
while True:
  ask=input("簽到或登錄資料？（簽到／登錄）(若要結束程式輸入「結束」)")
  if ask=="結束":
    print("結束")
    print("簽到名單：",(attendance_log))
    print(f"簽到人數：{len(attendance_log)}人")
    break

  #登錄功能
  elif ask=="登錄":
    while True:
      sid=input(f"輸入學號({ID_LENGTH}位數字)：").strip()
      seat=input(f"輸入班級座號({SEAT_LENGTH}位數字)：").strip()
      name=input("輸入姓名(不可為空白或數字)：")
      if len(sid)!=ID_LENGTH or len(seat)!=SEAT_LENGTH or name=="" or not name.isalpha():
        print("學號或班級座號或姓名格式錯誤")
        print("請重新輸入")
        continue
      member_db[sid]=[sid,seat,name]
      print(f"{seat}{name}登錄成功")
      print(f"目前共有{len(member_db)}位團員")
      print("將回到主選單")
      break
  elif ask=="簽到":
  #簽到功能
    while True:
      check_id=input("輸入簽到人學號").strip()
      if check_id not in member_db:
        print("尚未登錄")
        print("請先登錄")
        break
      if check_id in attendance_log:
        already_info=attendance_log[check_id]
        print(f"{already_info[0]}{already_info[1]}已簽過（第一次時間為{already_info[2]}）")
        print("將回到主選單")
        break
      else:
        info=member_db[check_id]
        seat=info[1]
        name=info[2]
        time_now=datetime.now().strftime("%H:%M:%S")
        attendance_log[check_id]=[seat,name,time_now]
        print(f"{seat}{name}{time_now}簽到成功")
        print("將回到主選單")
        break
  else:
    print("輸入無效，請重新選擇")
    continue


import pandas as pd
from datetime import datetime
def save_to_excel(attendance_log, member_db):
    if not attendance_log:
        print("目前沒有簽到紀錄，無法產出 Excel。")
        return

    # 1. 整理資料：將字典轉換為清單格式
    data_list = []
    for sid, info in attendance_log.items():
        # info 內容為 [座號, 姓名, 簽到時間]
        data_list.append({
            "班級座號": info[0],
            "姓名": info[1],
            "簽到時間": info[2],
            "狀態": "已簽到"
        })

    # 2. 建立 DataFrame
    df = pd.DataFrame(data_list)

    # 3. 排序（依座號排序，讓報表更整齊）
    df = df.sort_values(by="班級座號")

    # 4. 產出 Excel 檔
    file_name = input("輸入檔案名稱")
    final_file_name=f"{file_name}.xlsx"
    df.to_excel(final_file_name, index=False, engine='openpyxl')

    print(f"\n--- 報表產出成功 ---")
    print(f"檔案名稱：{file_name}")
    print(f"總簽到人數：{len(df)} 人")

# 在主程式結束時呼叫：
save_to_excel(attendance_log, member_db)
```
